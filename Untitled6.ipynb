```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the data using your URL
url = 'https://raw.githubusercontent.com/Hemamalini-19/scm-tracker/refs/heads/main/po_tracking.csv'
df = pd.read_csv(url, parse_dates=['expected_delivery', 'actual_delivery'])

# 2. Calculate delays
df['delay_days'] = (df['actual_delivery'] - df['expected_delivery']).dt.days

# 3. Create a Visual Report
plt.figure(figsize=(10, 6))
sns.barplot(x='supplier', y='delay_days', data=df, palette='viridis')

# 4. Add labels for the SCM Lead to understand
plt.title('Average Supplier Delivery Delays (Days)', fontsize=14)
plt.xlabel('Supplier Name', fontsize=12)
plt.ylabel('Days Delayed', fontsize=12)
plt.axhline(0, color='black', linewidth=0.8) # Baseline for on-time

plt.show()
```
